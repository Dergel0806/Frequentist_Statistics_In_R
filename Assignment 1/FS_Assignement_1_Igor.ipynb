```r
traits <- c('traitX', 'traitY', 'traitZ')
items <- c('itemA','itemB', 'itemC', 'itemD', 'itemE', 'itemF', 'itemG')
```

## Q1

```r
for(trait in traits){
    for(item in items){
        ds1wilcox <- wilcox.test(ds1[[trait]] ~ ds1[[item]], paired = FALSE, exact = FALSE) 
        cat(trait, ' | ', item, ' | ', ds1wilcox$p.value)
        cat('\n')
    }
}
```

Print only significant values:

```r
for(trait in traits){
    for(item in items){
        ds1wilcox <- wilcox.test(ds1[[trait]] ~ ds1[[item]], paired = FALSE, exact = FALSE) 
        if(ds1wilcox$p.value < 0.05){
            cat(trait, ' | ', item, ' | ', ds1wilcox$p.value)
            cat('\n')
        }
    }
}
```

```r
summary(xtabs(ds1[items]))
```

```r
chi.p <- data.frame(item1=c(),item2=c(), p=c())
for(i in 1:(length(items)-1)){
    for(j in (i+1):length(items)){
         chi.p <- rbind(chi.p, data.frame(item1=items[i], item2=items[j], 
                                          p=CrossTable(ds1[[items[i]]], ds1[[items[j]]], fisher = T, chisq = F, 
           expected = TRUE, sresid = FALSE) %>% 
        .$chisq %>%
        .$p.value))    
    }
}
```

```r
chi.p[] <- lapply(chi.p, as.character) 
# https://stackoverflow.com/questions/2851015/convert-data-frame-columns-from-factors-to-characters


for(i in 1:length(chi.p[,1])){
    cat(chi.p[i,1], ' | ', chi.p[i,2], ' | ', round(as.numeric(chi.p[i,3]), 5), '\n')
}
```

```r
if(nrow(chi.p[chi.p['p'] < 0.05,]) == 0){
   cat('No significant values!\n') 
} else {
    for(i in 1:nrow(chi.p[chi.p['p'] < 0.05,])){
        cat(chi.p[chi.p['p'] < 0.05,][i, 1], ' | ', 
            chi.p[chi.p['p'] < 0.05,][i, 2], ' | ', 
            as.numeric(chi.p[chi.p['p'] < 0.05,][i, 3]), '\n')
    }
}
```
